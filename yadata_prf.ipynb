# Генерируем данные

```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport

# 1. Генерируем данные
np.random.seed(0)
n_samples = 1000

# Генерируем числовые данные для разных признаков
data = {
    'Age': np.random.randint(18, 70, size=n_samples),  # возраст от 18 до 70
    'Salary': np.random.randint(30000, 120000, size=n_samples),  # зарплата от 30k до 120k
    'Score': np.random.normal(loc=75, scale=10, size=n_samples),  # баллы (нормальное распределение)
    'Gender': np.random.choice(['Male', 'Female'], size=n_samples)  # пол
}

df = pd.DataFrame(data)
```

```python
df.head()
```

## 2. Добавляем выбросы
Добавим выбросы в столбцы "Salary" и "Score"

```python
outliers = pd.DataFrame({
    'Age': np.random.randint(18, 70, size=5),
    'Salary': np.random.randint(150000, 200000, size=5),  # выбросы в зарплате
    'Score': np.random.randint(10, 30, size=5),  # выбросы в баллах
    'Gender': np.random.choice(['Male', 'Female'], size=5)
})
df = pd.concat([df, outliers], ignore_index=True)
```

## 3. Добавляем пустые значения

```python
df.loc[::50, 'Salary'] = np.nan  # добавляем пропуски в Salary через каждые 50 строк
df.loc[::100, 'Score'] = np.nan  # добавляем пропуски в Score через каждые 100 строк
```

## 4. Анализируем данные с помощью yadataprofiling

```python
profile = ProfileReport(df, title="Data Profiling Report", explorative=True)

# Сохраняем отчет в HTML файл
profile.to_file("data_profile_report.html")
```

# Улучшаем качество данных

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from fancyimpute import IterativeImputer
```

## 1. Обработка пропусков с использованием SimpleImputer (заполнение медианой)

```python
imputer = SimpleImputer(strategy='median')
df[['Age', 'Salary', 'Score']] = imputer.fit_transform(df[['Age', 'Salary', 'Score']])
```

## 2. Обработка выбросов: ограничиваем значения столбца Salary
Для обработки выбросов в столбце Salary используется метод клиппинга, который ограничивает значения в столбце, чтобы они не превышали определенную границу (в данном случае 120,000).

```python
df['Salary'] = df['Salary'].clip(upper=120000)
```

## 3. Использование MICE для более сложной импутации

```python
mice_imputer = IterativeImputer()
df[['Age', 'Salary', 'Score']] = pd.DataFrame(mice_imputer.fit_transform(df[['Age', 'Salary', 'Score']]), columns=['Age', 'Salary', 'Score'])

# Проверим результат
print(df.head())
```

```python
profile = ProfileReport(df, title="Data Profiling Report", explorative=True)
```
