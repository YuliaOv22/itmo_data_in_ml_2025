```python
import pandas as pd
import ast
```

# Агрегация данных

### РИА Новости

```python
ria_news_v1 = pd.read_csv('data/ria_news_archive_full_v1.csv', sep=',')
ria_news_v1.shape
```

```python
ria_news_v1 = ria_news_v1.drop('page', axis=1)
ria_news_v1.head()
```

```python
ria_news_v2 = pd.read_csv('data/ria_news_archive_full_v2.csv', sep=',')
ria_news_v2.shape
```

```python
ria_news_v2 = ria_news_v2.drop('page', axis=1)
ria_news_v2.head()
```

### Lenta.ru

```python
lenta_news = pd.read_csv('data/lenta_news.csv', sep=',')

data_list = lenta_news['0'].tolist()

# Функция парсинга словаря в строке
def safe_literal_eval(s):
    return ast.literal_eval(str(s))

parsed_data = [safe_literal_eval(item) for item in data_list]
lenta_news = pd.DataFrame([x for x in parsed_data if x])

# Обработка столбцов со списками
if 'full_text' in lenta_news.columns:
    lenta_news['full_text'] = lenta_news['full_text'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
    
if 'tags' in lenta_news.columns:
    lenta_news['tags'] = lenta_news['tags'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

lenta_news['target'] = 'True'
lenta_news.shape
```

```python
lenta_news = lenta_news.drop('page', axis=1)
lenta_news.head()
```

### Фейковые новости из Telegram-каналов

```python
fake_news = pd.read_csv('data/tgstat_posts_fake.csv', sep=',')
fake_news['tags'] = None
fake_news['target'] = 'False'
fake_news.shape
```

```python
fake_news = fake_news.drop('author', axis=1)
fake_news.head()
```

### Объединение данных в один датасет

```python
combined = pd.concat([ria_news_v1, ria_news_v2, lenta_news, fake_news], axis=0)
combined
```

```python
combined.to_csv('data/real_fake_news_dataset.csv', encoding='utf-8-sig')
```
